```python
from src.cfbd_client import configuration
import cfbd
import pandas as pd
import os

os.makedirs("../Data/Raw", exist_ok=True)

with cfbd.ApiClient(configuration) as api_client:
    stats_api = cfbd.StatsApi(api_client)

    all_stats = []

    for year in range(2000, 2025):
        print(f"Fetching team stats for {year}...")
        try:
            stats = stats_api.get_team_stats(year=year)
            df_stats = pd.DataFrame([s.to_dict() for s in stats])
            df_stats["season"] = year
            all_stats.append(df_stats)
        except Exception as e:
            print(f"⚠️ Failed for {year}: {e}")

df_team_stats = pd.concat([df for df in all_stats if not df.empty], ignore_index=True)
print(f"Combined team stats shape: {df_team_stats.shape}")

df_team_stats.to_csv("../Data/Raw/team_stats_2000_2025.csv", index=False)

df_team_stats.head()
```
