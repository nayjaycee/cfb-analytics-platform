## Multi-Year Data Pull


```python
from src.cfbd_client import get_games_by_year
from src.data_cleaning import clean_games
import pandas as pd
import os
from time import sleep

raw_path = "../Data/Raw"
os.makedirs(raw_path, exist_ok=True)

START_YEAR = 2000
END_YEAR = 2025

all_years = []

for year in range(START_YEAR, END_YEAR + 1):
    print(f"Fetching {year} season...")

    try:
        df = get_games_by_year(year)
        df = clean_games(df)
        df["season"] = year

        output_path = os.path.join(raw_path, f"games_{year}.csv")
        df.to_csv(output_path, index=False)

        all_years.append(df)

        sleep(0.5)

    except Exception as e:
        print(f"⚠️ Failed for {year}: {e}")

df_all = pd.concat(all_years, ignore_index=True)
print(f"\nCombined dataset shape: {df_all.shape}")

df_all.to_csv(os.path.join(raw_path, "games_all_2000_2025.csv"), index=False)

df_all.head()
```
